根据test.txt中的 第一列（原来的remap_id, 包含全部user的排序）， 去到user_list中 对照相应remap_id，重新生成一份user_list， 即只保留当前train.txt 中含有的user original id, 并按照剩余的user org id重新生成remap_id

```python
from pandas.core.frame import DataFrame
from pandasql import sqldf
import pandas as pd
```

```python
left_user_remap_id = []
with open("test.txt") as f:
    for l in f.readlines():
            l = l.strip('\n').split(' ')
            left_user_remap_id.append(int(l[0])) 

```

```python
left_user_remap_id = {"part_remap_id_user": left_user_remap_id}
left_user_remap_id = pd.DataFrame(left_user_remap_id)
```

```python
left_user_remap_id
```

```python
customer_list = pd.read_table('user_list.txt',sep='\t')  
customer_list
```

```python
left_user_list = sqldf("""SELECT org_id, part_remap_id_user 
        FROM customer_list as c
        INNER JOIN left_user_remap_id as l
        on c.remap_id = l.part_remap_id_user
    """)
print(left_user_list)
```

```python
left_user_list.columns = ['org_id','remap_id']
```

```python
left_user_list
```

```python
left_user_list['remap_id'] = left_user_list.index
left_user_list
```

```python
left_user_list.to_csv(r'user_list_1.txt', index=None, sep='\t', mode='a')
```
